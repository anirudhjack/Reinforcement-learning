$
\textbf{$\epsilon$ -greedy improvement over π}:Given\;policy\;we\;select\;a\;action\;which\;maximizes\;the\;reward\;with\;a\;probability\;1-\epsilon.
$

$
\textbf{Proof}:Consider\;the\;Action\;value\;function\;of\;a\;state\;s:q_{\pi}(s,\pi^{'}(s))\\
\implies We\;have\;q_{\pi}(s,\pi^{'}(s))=\sum_{a\in A}\pi^{'}(\frac{a}{s})q_{\pi}(s,a).\\
=\epsilon/m.\sum_{a\in A}q_{\pi}(s,a)+(1-\epsilon).\sum_{a\in A}max(q_{\pi}(s,a)))\\
Since\;the\;max\;operation\;is\;greater\;than\;equal\;to\;an\;arbitrary\;weighted\;sum.\\
\geq \epsilon/m.\sum_{a\in A}q_{\pi}(s,a)+(1-\epsilon ).\sum_{a\in A}\frac{\pi(a/s)-\epsilon/m}{1-\epsilon}.q_{\pi}(s,a).\\
=\sum_{a\in A}\pi(a/s)q_{\pi}(s,a)\\
=V_{\pi}(s)\\
\therefore Therefore\;from\;policy\;improvement\;theorem\;V_{\pi^{'}}(s)\geq V_{\pi}(s).  
$

```
import numpy as np
import matplotlib.pyplot as plt
import random
import math
```

$
(a)Init\;function\;and\;neccessary\;functions\;for\;playing\;Pac-man\;game.
$

```
class Pac_man:
  def __init__(self,size,food_pallets):
    self.size=size
    self.board=list()
    self.food_positions=list()
    self.reward=np.zeros([size+2,size+2])
    self.score=0
    self.food_pallets=food_pallets

    #Create board
    for i in range(size+2):
      row=list()
      for j in range(size+2):
        if i==0 or j==0 or i==size+1 or j==size+1:
          row.append('W')
        else:
          row.append('-')
      self.board.append(row)    

    #Initialize the position of pacman
    x=random.randint(1,size-1)
    y=random.randint(1,size-1)
    self.board[x][y]='P'

    #Initialize the position of ghost
    while(True):
      if random.randint(0,1)==1:
        self.board[x][size]='G'
        break
      else:
        self.board[size][y]='G'
        break

    #Place food_pallets
    food_pallet=self.food_pallets
    while(food_pallet):
      x=random.randint(1,size)
      y=random.randint(1,size)
      if self.board[x][y]!='F' and self.board[x][y]!='P' and self.board[x][y]!='G':
        self.board[x][y]='F'
        self.food_positions.append([x,y])
        food_pallet-=1

    for i in range(size+2):
      for j in range(size+2):
        if self.board[i][j]=='-' or self.board[i][j]=='P':
          self.reward[i][j]=0
        elif self.board[i][j]=='F':
          self.reward[i][j]=5
        elif self.board[i][j]=='G':
          self.reward[i][j]=-5
        else:
          self.reward[i][j]=-20

  def print(self):
    print(np.array(self.board))

  def get_ghost_position(self):
    for i in range(1,self.size+1):
      for j in range(1,self.size+1):
        if self.board[i][j]=='G':
          return [i,j]

  def get_pacman_position(self):
    for i in range(self.size+2):
      for j in range(self.size+2):
        if self.board[i][j]=='P':
          return [i,j]

  def act(self,move):
    prev_pos=self.get_pacman_position()
    if self.board[move[0]][move[1]]=='F':
      self.food_positions.remove(move)
      self.score+=1
      self.food_pallets-=1
    self.board[move[0]][move[1]]='P'
    if prev_pos!=move:
      self.board[prev_pos[0]][prev_pos[1]]='-'
      self.reward[prev_pos[0]][prev_pos[1]]=0
    
    if self.food_pallets==0:
      new_food_pallets=random.randint(0,self.size)
      self.food_pallets=new_food_pallets
      while(new_food_pallets):
        x=random.randint(1,self.size)
        y=random.randint(1,self.size)
        if self.board[x][y]!='F' and self.board[x][y]!='P' and self.board[x][y]!='G':
          self.board[x][y]='F'
          self.food_positions.append([x,y])
          self.reward[x][y]=5
          new_food_pallets-=1

  def check_loss(self,move):
    if self.board[move[0]][move[1]]=='W':
      return 1
    elif self.board[move[0]][move[1]]=='G':
      return 2 
    else:
      return 0
  
  def epsilon_greedy(self,epsilon,Q_table):
    pacman_position=self.get_pacman_position()
    current_max=-math.inf
    move=0
    if random.uniform(0,1)<epsilon:
      move=random.randint(0,3)
    else:
      current_max=Q_table[pacman_position[0]][pacman_position[1]][0]
      for i in range(4):
        if current_max<Q_table[pacman_position[0]][pacman_position[1]][i]:
          current_max=Q_table[pacman_position[0]][pacman_position[1]][i]
          move=i

    return (move,current_max)

  def ghost_move(self):
    position=self.get_ghost_position()
    random_move_x=0
    random_move_y=0
    options=[-1,1]

    if random.randint(0,1)==1:
      random_move_x=random.choice(options)
    else:
      random_move_y=random.choice(options)

    new_pos=[-1,1]
    if self.board[position[0]+random_move_x][position[1]+random_move_y]!='W':
      self.board[position[0]+random_move_x][position[1]+random_move_y]='G'
      self.reward[position[0]+random_move_x][position[1]+random_move_y]=-5
      if position in self.food_positions:
        self.board[position[0]][position[1]]='F'
        self.reward[position[0]][position[1]]=5
      else:
        self.board[position[0]][position[1]]='-'
        self.reward[position[0]][position[1]]=0
    else:
      if position[0]+random_move_x==0:
        new_pos[0]=self.size
        new_pos[1]=position[1]
      elif position[0]+random_move_x==self.size+1:
        new_pos[0]=1
        new_pos[1]=position[1]
      
      if position[1]+random_move_y==0:
        new_pos[1]=self.size
        new_pos[0]=position[0]
      elif position[1]+random_move_y==self.size+1:
        new_pos[1]=1
        new_pos[0]=position[0]
      
      self.board[new_pos[0]][new_pos[1]]='G'
      self.reward[new_pos[0]][new_pos[1]]=-5
      if position in self.food_positions:
        self.board[position[0]][position[1]]='F'
        self.reward[position[0]][position[1]]=5
      else:
        self.board[position[0]][position[1]]='-'
        self.reward[position[0]][position[1]]=0


```

$
Creating\;Q-table\;for\;tabular\;Q-learning\;and\;SARSA.
$

```
def initialize_Q_table(size):
  Q_table=list()
  for i in range(size+2):
    row=list()
    for j in range(size+2):
      udrls=list()
      for k in range(4):
        udrls.append(random.random())
      row.append(udrls)
    Q_table.append(row)

  return Q_table 
```

$
Developing\;a\;Q-learning\;agent\;for\;playing\;Pac-man\;game.\\
\rightarrow Q(s_{t},a_{t})=Q(s_{t},a_{t})*(1-\alpha)+\alpha*(r_{t+1}+\gamma.max_{a'}Q(s_{t+1},a'))
$

```
def Q_learning_pacman(iterations,alpha,gamma,epsilon,size,food_pallets):
  Q_sa=initialize_Q_table(size)
  for i in range(iterations):
    game=Pac_man(size=size,
                food_pallets=food_pallets)

    while(True):
      s_t=game.get_pacman_position()
      move_pac=game.epsilon_greedy(epsilon,Q_sa)[0]
      move=[s_t[0],s_t[1]]
      
      if move_pac==0:
        move[0]-=1
      elif move_pac==1:
        move[0]+=1
      elif move_pac==2:
        move[1]+=1
      elif move_pac==3:
        move[1]-=1
      
      if game.check_loss(move)!=0:
        game.act(move)
        s_tplus1=game.get_pacman_position()
        Q_sa[s_t[0]][s_t[1]][move_pac]=(1-alpha)*Q_sa[s_t[0]][s_t[1]][move_pac]+alpha*(game.reward[s_t[0]][s_t[1]]+gamma*Q_sa[s_tplus1[0]][s_tplus1[1]][0])
        # print(game.score)
        break

      Q_sa[s_t[0]][s_t[1]][move_pac]=(1-alpha)*Q_sa[s_t[0]][s_t[1]][move_pac]+alpha*(game.reward[s_t[0]][s_t[1]]+gamma*game.epsilon_greedy(-math.inf,Q_sa)[1])
      game.act(move)
      
      new_s_t=game.get_pacman_position()
      game.ghost_move()
      ghost_position=game.get_ghost_position()
      if ghost_position==new_s_t:
        # print(game.score)
        break
    if i%(iterations/10)==0:
      print('%d epochs are completed'%(i+iterations/10))

  print('Training of Q-learning agent is completed')
  return Q_sa
```

```
def SARSA_pacman(iterations,alpha,gamma,epsilon,size,food_pallets):
  Q_sa=initialize_Q_table(size)
  for i in range(iterations):
    game=Pac_man(size=size,
                food_pallets=food_pallets)
    
    move_pac=-1
    while(True):
      s_t=game.get_pacman_position()
      move=[s_t[0],s_t[1]]
      
      if move_pac==0:
        move[0]-=1
      elif move_pac==1:
        move[0]+=1
      elif move_pac==2:
        move[1]+=1
      elif move_pac==3:
        move[1]-=1
      
      if game.check_loss(move)!=0:
        game.act(move)
        s_tplus1=game.get_pacman_position()
        Q_sa[s_t[0]][s_t[1]][move_pac]=(1-alpha)*Q_sa[s_t[0]][s_t[1]][move_pac]+alpha*(game.reward[s_t[0]][s_t[1]]+gamma*Q_sa[s_tplus1[0]][s_tplus1[1]][0])
        # print(game.score)
        break

      game.act(move)
      Q_sa[s_t[0]][s_t[1]][move_pac]=(1-alpha)*Q_sa[s_t[0]][s_t[1]][move_pac]+alpha*(game.reward[s_t[0]][s_t[1]]+gamma*game.epsilon_greedy(-math.inf,Q_sa)[1])
      move_pac=game.epsilon_greedy(epsilon,Q_sa)[0]
      
      new_s_t=game.get_pacman_position()
      game.ghost_move()
      ghost_position=game.get_ghost_position()
      if ghost_position==new_s_t:
        # print(game.score)
        break
    if i%(iterations/10)==0:
      print('%d epochs are completed'%(i+iterations/10))

  print('Training of SARSA agent is completed')
  return Q_sa
```

```
class Tic_tac_toe:
  def __init__(self):
    self.board =np.array([['-', '-', '-'], ['-', '-', '-'], ['-', '-', '-']])

  def act(self,move):
    if self.board[move[0]][move[1]]=='-':
      self.board[move[0]][move[1]]='X'
    
  def print(self):
    print(self.board)

  def get_valid_coordinates(self):
    valid_coordinates=list()
    for i in range(np.shape(self.board)[0]):
      for j in range(np.shape(self.board)[1]):
        if self.board[i][j]=='-':
          valid_coordinates.append([i,j])
    return valid_coordinates

  def random_agent(self):
    valid_places=self.get_valid_coordinates()
    random_move=random.choice(valid_places)
    self.board[random_move[0]][random_move[1]]='O'
    return random_move

  def full(self):
    for i in range(np.shape(self.board)[0]):
      for j in range(np.shape(self.board)[1]):
        if self.board[i][j]=='-':
          return 0
    return 1

  def check_win(self,position,mark):
    if position==[0,0]:
      if (self.board[0][1]==mark and self.board[0][2]==mark) or (self.board[1][0]==mark and self.board[2][0]==mark) or (self.board[1][1]==mark and self.board[2][2]==mark):
        return 1
    elif position==[0,1]:
      if (self.board[0][0]==mark and self.board[0][2]==mark) or (self.board[1][1]==mark and self.board[2][1]==mark):
        return 1
    elif position==[0,2]:
      if (self.board[0][0]==mark and self.board[0][1]==mark) or (self.board[1][2]==mark and self.board[2][2]==mark) or (self.board[1][1]==mark and self.board[2][0]==mark):
        return 1
    elif position==[1,0]:
      if (self.board[0][0]==mark and self.board[2][0]==mark) or (self.board[1][1]==mark and self.board[1][2]==mark):
        return 1
    elif position==[1,1]:
      if (self.board[0][0]==mark and self.board[2][2]==mark) or (self.board[2][0]==mark and self.board[0][2]==mark) or (self.board[1][0]==mark and self.board[1][2]==mark) or (self.board[0][1]==mark and self.board[2][1]==mark):
        return 1
    elif position==[1,2]:
      if (self.board[1][0]==mark and self.board[1][1]==mark) or (self.board[0][2]==mark and self.board[2][2]==mark):
        return 1
    if position==[2,0]:
      if (self.board[0][0]==mark and self.board[1][0]==mark) or (self.board[2][1]==mark and self.board[2][2]==mark) or (self.board[1][1]==mark and self.board[0][2]==mark):
        return 1
    elif position==[2,1]:
      if (self.board[2][0]==mark and self.board[2][2]==mark) or (self.board[1][1]==mark and self.board[0][1]==mark):
        return 1
    elif position==[2,2]:
      if (self.board[2][0]==mark and self.board[2][1]==mark) or (self.board[0][2]==mark and self.board[1][2]==mark) or (self.board[1][1]==mark and self.board[0][0]==mark):
        return 1
    else :
      return 0


  def safe_agent(self):
    valid_places=self.get_valid_coordinates()
    for pos in valid_places:
      if self.check_win(position=pos,mark='X')==1 :#Check for the win
        self.board[pos[0]][pos[1]]='O'
        return pos
      elif self.check_win(position=pos,mark='O')==1:#Check for the blocking move
        self.board[pos[0]][pos[1]]='O'
        return pos
    
    random_move=random.choice(valid_places)
    self.board[random_move[0]][random_move[1]]='O'#Else give random move
    return random_move

  def epsilon_greedy(self,epsilon):
    valid_places=self.get_valid_coordinates()
    current_max=-math.inf
    move=[0,0]
    if random.uniform(0,1)<epsilon:
      move=random.choice(valid_places)
    else:
      string=''.join([str(mark) for mark in self.board.flatten()])
      current_max=Q_table[string][0]
      for position in valid_places:
        if current_max<Q_table[string][3*position[0]+position[1]]:
          current_max=Q_table[string][3*position[0]+position[1]]
          move=position

    return (move,current_max)


  
```

```
def count_result(trails,agent,epsilon):
  won,draw,losses=0,0,0

  for i in range(trails):
    trail_game=Tic_tac_toe()
    toss=random.randint(0,1)

    if toss==1:
      if agent=="random":
         trail_game.random_agent()
      elif agent=="safe":
         trail_game.safe_agent()
    
    while(True):
      move_X=trail_game.epsilon_greedy(epsilon)[0]
      trail_game.act(move_X)
      #Blocking move : 
      if trail_game.check_win(move_X,'X')==1:
        won+=1
        break
      elif trail_game.full()==1:
        draw+=1
        break

      #safe or random move : 
      if agent=="random":
        move_O=trail_game.random_agent()

      elif agent=="safe":
        move_O=trail_game.safe_agent()

      else:
        if np.random.randint(0,1)==1:
          move_O=trail_game.random_agent()
        else:
          move_O=trail_game.safe_agent()


      if trail_game.check_win(move_O,'O')==1:
        losses+=1
        break
      elif trail_game.full()==1:
        draw+=1
        break

  return [won,draw,losses]
```

```
def Q_learning(iterations,opponent,alpha,gamma,epsilon):
  backtrack_qtable(board=np.array([['-', '-', '-'], ['-', '-', '-'], ['-', '-', '-']])
                    ,length=0)
  
  wins=list()
  wins_random=list()
  wins_safe=list()
  epochs=list()
  for i in range(iterations):
    game=Tic_tac_toe()
    toss=random.randint(0,1)
    if toss==1:
      move=game.random_agent()

    while(True):
      s_t=''.join([str(box) for box in game.board.flatten()])
      move_X=game.epsilon_greedy(epsilon)[0]
      game.act(move_X)

      reward=0
      if game.check_win(move_X,'X')==1:
        s_tplus1=''.join([str(box) for box in game.board.flatten()])
        Q_table[s_tplus1]=[math.inf]*9
        reward=1
        Q_table[s_t][3*move_X[0]+move_X[1]]=(1-alpha)*Q_table[s_t][3*move_X[0]+move_X[1]]+alpha*(reward+gamma*Q_table[s_tplus1][0])
        break
      elif game.full()==1:
        s_tplus1=''.join([str(box) for box in game.board.flatten()])
        reward=0.5
        Q_table[s_t][3*move_X[0]+move_X[1]]=(1-alpha)*Q_table[s_t][3*move_X[0]+move_X[1]]+alpha*(reward+gamma*Q_table[s_tplus1][0])
        break

      if opponent=='random':
        move_O=game.random_agent()
      elif opponent=='safe':
        move_O=game.safe_agent()
      else :
        random_select=random.randint(0,1)
        if random_select==1:
          move_O=game.random_agent()
        else :
          move_O=game.safe_agent()
      
      if game.check_win(move_O,'O')==1:
        s_tplus1=''.join([str(box) for box in game.board.flatten()])
        Q_table[s_tplus1]=[0]*9
        reward=-0.5
        Q_table[s_t][3*move_O[0]+move_O[1]]=(1-alpha)*Q_table[s_t][3*move_O[0]+move_O[1]]+alpha*(reward+gamma*Q_table[s_tplus1][0])
        break
      elif game.full()==1:
        s_tplus1=''.join([str(box) for box in game.board.flatten()])
        reward=0.5
        Q_table[s_t][3*move_O[0]+move_O[1]]=(1-alpha)*Q_table[s_t][3*move_O[0]+move_O[1]]+alpha*(reward+gamma*Q_table[s_tplus1][0])
        break

      Q_table[s_t][3*move_X[0]+move_X[1]]=(1-alpha)*Q_table[s_t][3*move_X[0]+move_X[1]]+alpha*(reward+gamma*game.epsilon_greedy(epsilon=-math.inf)[1])

    if i%200==0:
      epochs.append(i)
      if opponent=='random_select':
        wins_random.append(count_result(trails=100,
                                        agent='random',
                                        epsilon=0)[0])
        
        wins_safe.append(count_result(trails=100,
                                      agent='safe',
                                      epsilon=0)[0])
      else:
        wins.append(count_result(trails=100 ,
                                 agent=opponent,
                                 epsilon=0)[0])   

  if opponent=='random_select':
    plt.plot(epochs,wins_random,label='random_player')
    plt.plot(epochs,wins_safe,label='safe_player')
    plt.legend()
    plt.xlabel('epochs')
    plt.ylabel('wins')
    plt.show()
  else:
    plt.plot(epochs,wins,label=opponent+' player')
    plt.legend()
    plt.xlabel('epochs')
    plt.ylabel('wins')
    plt.show()

  random_result=count_result(10000 ,"random",epsilon=0)
  safe_result=count_result(10000 ,"safe",epsilon=0)

  print("No of games won against random player :%d"%(random_result[0]))
  print("No of games drawn against random player :%d"%(random_result[1]))
  print("No of games lost against random player :%d"%(random_result[2]))
  print("-------------------------------------------------------------")
  print("No of games won against safe player :%d"%(safe_result[0]))
  print("No of games drawn against safe player :%d"%(safe_result[1]))
  print("No of games lost against safe player :%d"%(safe_result[2]))
```
